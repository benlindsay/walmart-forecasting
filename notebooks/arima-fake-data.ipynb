```python
from dotenv import find_dotenv
from os.path import dirname
from os.path import join
import matplotlib.pyplot as plt
import numpy as np

project_dir = dirname(find_dotenv())

plt.style.use(join(project_dir, 'big-darkgrid.mplstyle'))
```

```python
def generate(size, mu, sigma, phi_1, phi_2):
    data = np.zeros(size)
    for i in range(size):
        data[i] = mu
        if i > 0:
            data[i] += phi_1 * data[i-1]
        if i > 1:
            data[i] += phi_2 * data[i-2]
        data[i] += sigma * np.random.normal()
    return data 
```

```python
from itertools import product
phi_1_vals = [-0.1, 0, 0.1]
phi_2_vals = [-0.1, 0, 0.1]
for phi_1, phi_2 in product(phi_1_vals, phi_2_vals):
    arr = generate(200, 5, 0.1, phi_1, phi_2)
    plt.plot(arr, label=r'$\phi_1={}, \phi_2={}$'.format(phi_1, phi_2), alpha=0.5)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
```
